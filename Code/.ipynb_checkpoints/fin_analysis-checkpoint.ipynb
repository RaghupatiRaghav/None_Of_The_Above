```python
#creating variable to determine the position on evm of independent candidates. So the independent with name starting from A should feature as the first position on evm with respect to other independent candidates
#shifts uni_const_code and year variables in (i+1)th row and then checks if it is the same as the ith row
finding_salient['uni_const_code_check'] = finding_salient.uni_const_code.eq(finding_salient.uni_const_code.shift(periods = 1))
finding_salient['year_check'] = finding_salient.Year.eq(finding_salient.Year.shift(periods = 1))
finding_salient['final_check'] = (finding_salient['uni_const_code_check'] & finding_salient['year_check'])
#the last code gives false if either of the checks are false. It will yield true iff both the checks are true
```

```python
#position_on_evm_ind is the variable that determines the position of the independent candidate and NOTA on the evm with respect to each other
s = finding_salient["final_check"].eq(False).cumsum()
finding_salient['position_on_evm_ind'] = s.groupby(s).cumcount() + 1
```

## **Regression Analysis**

```python
#subsetting the dataframe
df_reg = finding_salient[['index','uni_const_code','Year','pre_post_nota_ind','n_ind_w_nota','zi','position_on_evm_ind','position_on_evm_ind_lstbut2mrkr']][(finding_salient['n_ind_w_nota'].isin([2,3,4,5,6])) & (finding_salient['position_on_evm_ind_lstbut2mrkr'].isin([12345678,1234567,123456]))]
df_reg.head()
```

```python
#makes dummy attribute for last position
df_reg['last_position_dummy'] = 0
df_reg.loc[df_reg['position_on_evm_ind_lstbut2mrkr']==12345678, 'last_position_dummy'] = 1
df_reg.drop(columns = ['index'], inplace=True)
```

```python
#qc
df_reg[['position_on_evm_ind_lstbut2mrkr','last_position_dummy','pre_post_nota_ind']][(df_reg['position_on_evm_ind_lstbut2mrkr']!=12345678)&(df_reg['last_position_dummy']==1)]
#pass
```

```python
for_reg = df_reg[['n_ind_w_nota','last_position_dummy','pre_post_nota_ind','zi','position_on_evm_ind']]
for_reg.rename(columns = {'zi':'DPSi'}, inplace = True)
```

```python
for_reg['int_lastposition_prepostind'] = for_reg['last_position_dummy']*for_reg['pre_post_nota_ind']
```

```python
#Running a multiple linear regression using Statsmodels library
import statsmodels.api as sm
from statsmodels.iolib.summary2 import summary_col
X = for_reg[['n_ind_w_nota','last_position_dummy','position_on_evm_ind']]
X = sm.add_constant(X)
Y = for_reg[['DPSi']]
model = sm.OLS(Y,X,  M=sm.robust.norms.AndrewWave()).fit(cov_type='HC3')
print_model = model.summary()
print(print_model)
```

```python
res = summary_col([model])
print(res)
```

```python
with open(parent_dir+r'\Analysis\ttest_reg.csv', 'w') as fh:
    fh.write(print_model.as_csv())

with open(parent_dir+r'\Analysis\ttest_reg.txt', 'w') as fh:
    fh.write(print_model.as_text())
```
