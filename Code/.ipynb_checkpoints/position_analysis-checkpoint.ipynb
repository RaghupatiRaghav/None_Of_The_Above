```python
#QC
#we included only those constituencies that have independents or NOTA
#there will be some that don't have independents. checking how many const we lost pre and post nota due to this filter

# step1: making a party indicator. If party then 1, if independent or nota then 0
analysis_df['index'] = analysis_df.index
analysis_df['ind_ind'] = analysis_df['index'].apply(lambda i: 1 if analysis_df['Party'].iloc[i]=='IND' else 0)
```

```python
#for each constituency, arranging independents by alphabetical order and placing NOTA at the end of this list in post_nota period
finding_salient.sort_values(by = ['gen/state','state_code_new','Constituency_No','Year','Party','Candidate'], inplace = True)
finding_salient.reset_index(inplace = True)
```

```python
#QC
print(finding_salient.shape)
finding_salient[['uni_const_code','Year','Party','Candidate']][(finding_salient['uni_const_code']=='0|23|1') & (finding_salient['Year'].isin([2009,2014]))]
#pass
```

```python
#creating variable to determine the position on evm of independent candidates. So the independent with name starting from A should feature as the first position on evm with respect to other independent candidates
#shifts uni_const_code and year variables in (i+1)th row and then checks if it is the same as the ith row
finding_salient['uni_const_code_check'] = finding_salient.uni_const_code.eq(finding_salient.uni_const_code.shift(periods = 1))
finding_salient['year_check'] = finding_salient.Year.eq(finding_salient.Year.shift(periods = 1))
finding_salient['final_check'] = (finding_salient['uni_const_code_check'] & finding_salient['year_check'])
#the last code gives false if either of the checks are false. It will yield true iff both the checks are true
```

```python
#QC
finding_salient[['final_check','uni_const_code_check','year_check']].head(15)
#pass
```

```python
#position_on_evm_ind is the variable that determines the position of the independent candidate and NOTA on the evm with respect to each other
s = finding_salient["final_check"].eq(False).cumsum()
finding_salient['position_on_evm_ind'] = s.groupby(s).cumcount() + 1
```

## **Making the Zi score**    

```python
finding_salient.drop(columns = ['level_0'], inplace = True)
```

```python
#making the index column again
finding_salient.drop(columns = ['index'], inplace = True)
finding_salient['index'] = finding_salient.index
```

```python
#renaming cum vote share column to reflect that it is percentage, not just vote share
finding_salient.rename(columns = {'cum_vote_share_ind_nota':'cum_voteshare_perc_ind_nota'}, inplace = True)
```

```python
#making the statistic
finding_salient['zi'] = finding_salient['index'].apply(lambda i: (finding_salient['Vote_Share_Percentage'].iloc[i]/(finding_salient['cum_voteshare_perc_ind_nota'].iloc[i])) - (1/finding_salient['n_ind_w_nota'].iloc[i]))
```

```python
#QC
finding_salient[['index','uni_const_code','Year','Vote_Share_Percentage','cum_voteshare_perc_ind_nota','n_ind_w_nota','zi']].head()
#pass
```

## **Perspective -1 : same number of independents across elections, analyzing their vote shares by position**

```python
#arranging with_NOTA dataset as per our requirements
with_NOTA.sort_values(by = ['uni_const_code','Year'], inplace = True)
with_NOTA.reset_index(inplace = True)
with_NOTA.drop(columns = ['index'], inplace = True)
with_NOTA['index'] = with_NOTA.index
```

```python
#QC
print(with_NOTA.shape)
with_NOTA.head()
#pass
```

```python
#creating same_n_ind where if the number of ind candidates are the same in a constituency in subsequent elections, the indicator will yield 1 else 0
with_NOTA['same_n_ind'] = with_NOTA['index'].apply(lambda i: 1 if ((with_NOTA['uni_const_code'].iloc[i]==with_NOTA['uni_const_code'].iloc[i-1])&(with_NOTA['n_ind_w_nota'].iloc[i]==with_NOTA['n_ind_w_nota'].iloc[i-1])) else 0)

#shifting indicator 1 period up
with_NOTA['same_n_ind_shift'] = with_NOTA.same_n_ind.shift(periods = -1)
```

```python
#QC
print(with_NOTA.shape)
with_NOTA.tail(6)
#pass
```

```python
#QC
with_NOTA[with_NOTA['same_n_ind_shift'].isnull()]
#pass
```

```python
#summing up the two columns to get the final indicator. This has been done to ensure that we take not the only the election where n_cand_ind were same as the previous election but also the previous election
with_NOTA['same_n_ind_fin'] = with_NOTA['index'].apply(lambda i: (with_NOTA['same_n_ind'].iloc[i]+with_NOTA['same_n_ind_shift'].iloc[i]))
```

```python
#QC
with_NOTA[with_NOTA['same_n_ind_fin'].isin([2,3,4])]
#some elections will have the final indicator value of 2 when the number of candidates remain the same in three or more subsequent elections
#pass
```

```python
#making a dataset with constituencies of our interest
const_of_interest = with_NOTA[with_NOTA['same_n_ind_fin'].isin([1,2])]
```

```python
#QC
print(with_NOTA[with_NOTA['same_n_ind_fin'].isin([1,2])].shape)
print(const_of_interest.shape)
#pass
```

```python
#saving datasets as csv
with_NOTA.to_csv(parent_dir + r'Analysis\with_NOTA.csv', index = False)
const_of_interest.to_csv(parent_dir + r'Analysis\const_of_interest.csv', index = False)
```
