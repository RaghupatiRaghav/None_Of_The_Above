## Salient independent by position?

```python
#considering only independents and NOTA in cleanest_data.csv
finding_salient = analysis_df[(analysis_df['Party']=='IND')|(analysis_df['Party']=='NOTA')]
print(finding_salient.shape)
finding_salient.head()
```

```python
#QC
finding_salient[(finding_salient['Party']!='NOTA')&(finding_salient['Party']!='IND')]
#pass
```

```python
#for each constituency, arranging independents by alphabetical order and placing NOTA at the end of this list in post_nota period
finding_salient.sort_values(by = ['gen/state','state_code_new','Constituency_No','Year','Party','Candidate'], inplace = True)
```

```python
finding_salient.shape
#63,529 are total rows
```

```python
finding_salient['uni_const_code_check'] = finding_salient.uni_const_code.eq(finding_salient.uni_const_code.shift(periods = 1))
finding_salient['year_check'] = finding_salient.Year.eq(finding_salient.Year.shift(periods = 1))
```

```python
finding_salient.reset_index(inplace = True)
```

```python
finding_salient['index'] = finding_salient.index
finding_salient['position_on_evm'] = finding_salient['index'].apply(lambda i: 1 if ((finding_salient['uni_const_code_check'].iloc[i]==False) | (finding_salient['year_check'].iloc[i]==False)) else np.nan)
```

```python
finding_salient['position_on_evm_2'] = finding_salient['index'].apply(lambda i: finding_salient['position_on_evm'].cumsum(skipna=False) if (finding_salient['position_on_evm'].iloc[i]!=1) else 1)
```

```python
finding_salient.head()
```

```python
#reads csv
finding_salient = pd.read_csv(r"D:\MSE\Dissertation\Rebirth\NOTA_Github_publication\Analysis\finding_salient_excel.csv")
```

```python
#number of independents incl nota in each const in each year
with_NOTA = finding_salient[['uni_const_code', 'Year', 'gen/state']].groupby(['uni_const_code', 'Year']).count().reset_index()
with_NOTA.rename(columns = {'gen/state':'n_ind_w_nota'}, inplace = True)
with_NOTA.head()
```

```python
#left joining finding_salient with incl nota
finding_salient = pd.merge(finding_salient, with_NOTA, how='left', on=['uni_const_code','Year'])
```

```python
#QC
print(finding_salient.shape)
print(finding_salient[finding_salient['n_ind_w_nota'].isna()])
finding_salient[['n_ind_w_nota']][(finding_salient['uni_const_code']=='0|10|1') & (finding_salient['Year']==2019)]
#pass
```

```python
#adding new column to count no of independents excl nota
finding_salient['index'] = finding_salient.index
finding_salient['n_ind_wo_nota'] = finding_salient['index'].apply(lambda i: (finding_salient['n_ind_w_nota'].iloc[i]-1) if finding_salient['pre_post_nota_ind'].iloc[i]==1 else finding_salient['n_ind_w_nota'].iloc[i])
```

```python
#QC
finding_salient[['n_ind_w_nota','n_ind_wo_nota','pre_post_nota_ind','Year','uni_const_code']].head()
#pass
```

```python
#summing up vote of independent candidates including nota in each constituency election
vote_share_w_nota = finding_salient[['uni_const_code', 'Year', 'Vote_Share_Percentage']].groupby(['uni_const_code', 'Year']).sum().reset_index()
vote_share_w_nota.rename(columns = {'Vote_Share_Percentage':'cum_vote_share_ind_nota'}, inplace = True)
print(vote_share_w_nota.shape)
print(with_NOTA.shape)
vote_share_w_nota.head()
```

```python
#QC
print(0.55+.75+15.71+0.57+0.97+0.80+0.54+1.44+0.51)
finding_salient[['Year','Vote_Share_Percentage']][(finding_salient['uni_const_code']=='0|10|1')&(finding_salient['Year']==2009)]
#pass
```

```python
#left joining information on finding_salient
finding_salient = pd.merge(finding_salient, vote_share_w_nota, how='left', on=['uni_const_code','Year'])
print(finding_salient.shape)
```

```python
#QC
finding_salient[['Year','Vote_Share_Percentage','cum_vote_share_ind_nota']][(finding_salient['uni_const_code']=='0|10|1')&(finding_salient['Year']==2009)]
#pass
```

```python
#making a new column - expected vote share
finding_salient['expected_vote_share_w_nota'] = finding_salient['index'].apply(lambda i: finding_salient['cum_vote_share_ind_nota'].iloc[i]/finding_salient['n_ind_w_nota'].iloc[i])
finding_salient['expected_vote_share_wo_nota'] = finding_salient['index'].apply(lambda i: finding_salient['cum_vote_share_ind_nota'].iloc[i]/finding_salient['n_ind_wo_nota'].iloc[i])
```

```python
#QC
print(finding_salient.shape)
finding_salient[['index', 'uni_const_code', 'Year','cum_vote_share_ind_nota', 'n_ind_w_nota', 'n_ind_wo_nota','expected_vote_share_w_nota','expected_vote_share_wo_nota']].head()
```

```python
finding_salient.replace([np.inf, -np.inf], np.nan, inplace=True)
```

## Regression template

```python
#Trying out regression 
#Running a simple regression using Statsmodels library
import statsmodels.api as sm
X = finding_salient[['Position_on_evm']]
X = sm.add_constant(X)
Y = finding_salient[['Vote_Share_Percentage']]
model = sm.OLS(Y,X,  M=sm.robust.norms.AndrewWave()).fit()
print_model = model.summary()
print(print_model)
```
