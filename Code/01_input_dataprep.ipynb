```python
#imports general elections data using file search in rootdir
#filemust end with a .csv and there should be only one file in this directory
rootdir = parent_dir + r'Data\entire_data_general'
regex = re.compile('(.*csv$)')

for (root, dirs, files) in os.walk(rootdir, topdown = True):
    for file in files:
        if regex.match(file):
            gen = pd.read_csv(rootdir + '\\' + file)
```

```python
#imports general elections data using file search in rootdir
#filemust end with a .csv and multiple state files will be appended to state dataset which is initially created as an empty dataset with 50 columns
rootdir = parent_dir + r'Data\entire_data_assembly'
regex = re.compile('(.*csv$)')
state = pd.DataFrame(columns = ['Election_Type'
,'State_Name'
,'Assembly_No'
,'Constituency_No'
,'Year'
,'month'
,'Poll_No'
,'DelimID'
,'Position'
,'Candidate'
,'Sex'
,'Party'
,'Votes'
,'Candidate_Type'
,'Valid_Votes'
,'Electors'
,'Constituency_Name'
,'Constituency_Type'
,'Sub_Region'
,'N_Cand'
,'Turnout_Percentage'
,'Vote_Share_Percentage'
,'Deposit_Lost'
,'Margin'
,'Margin_Percentage'
,'ENOP'
,'pid'
,'Party_Type_TCPD'
,'Party_ID'
,'last_poll'
,'Contested'
,'Last_Party'
,'Last_Party_ID'
,'Last_Constituency_Name'
,'Same_Constituency'
,'Same_Party'
,'No_Terms'
,'Turncoat'
,'Incumbent'
,'Recontest'
,'Age'
,'District_Name'
,'PC_Name'
,'PC_No'
,'CandID'
,'MyNeta_education'
,'TCPD_Prof_Main'
,'TCPD_Prof_Main_Desc'
,'TCPD_Prof_Second'
,'TCPD_Prof_Second_Desc'])

for (root, dirs, files) in os.walk(rootdir, topdown = True):
    for file in files:
        if regex.match(file):
            read = pd.read_csv(rootdir + '\\' + file)
            state = state.append(read, ignore_index = True)
```

```python
#combines state and gen datasets
gen_state_combine = gen.append(state, ignore_index = True)
#Basic QC of the combined dataset
print(gen_state_combine.shape)
gen_state_combine.tail()

#the combined dataset has 159698 rows which = 25872+133826 (sum of gen and state datasets)
```

```python
#creates variable 'state_code_new' for census codes of states
census_code_df = pd.read_csv(parent_dir + r'\Data\ancillary_files\census_codes.csv')
print(census_code_df.shape)
census_code_df.head()
```

```python
#removes 'Region' column from census_code_df, renames 'Code' column as 'state_code_new'
census_code_df.drop(columns = 'Region', inplace = True)
census_code_df.rename(columns = {'Code':'state_code_new'}, inplace = True)
print(census_code_df.shape)
census_code_df.head()
```

```python
#matching state names in gen_state_combine and census_code_df for a left join
state_census = pd.DataFrame(census_code_df.States.unique(), columns = {'State_census_codefile'})
state_census = state_census.sort_values(by = 'State_census_codefile').reset_index()
state_census.drop(columns = 'index', inplace = True)
state_main = pd.DataFrame(gen_state_combine.State_Name.unique(), columns = {'State_maindf'})
state_main = state_main.sort_values(by = 'State_maindf').reset_index()
state_main.drop(columns = 'index', inplace = True)
name_match = pd.merge(state_census, state_main, left_index=True, right_index=True)
name_match
#all state names match
```

```python
#makes a left join of gen_state_combine with census_code_df
gen_state_combine = pd.merge(gen_state_combine, census_code_df, how = 'left', left_on = 'State_Name', right_on = 'States')
```

```python
#QC
qc_df = gen_state_combine[['State_Name','state_code_new','Turnout_Percentage']].groupby(by = ['State_Name','state_code_new']).mean().reset_index()
qc_df = qc_df.sort_values(by=['state_code_new']).reset_index()
qc_df.drop(columns =['index'], inplace = True)
qc_df
```

```python
#creates 'uni_const_id' variable as a unique identifier for constituency
gen_state_combine['Count'] = gen_state_combine.index
gen_state_combine['uni_const_code'] = gen_state_combine['Count'].apply(lambda x: str(gen_state_combine.iloc[x,:]['gen/state']) + '|' + str(gen_state_combine.iloc[x,:]['state_code_new']) + '|' + str(gen_state_combine.iloc[x,:]['Constituency_No']))
```

```python
#creates 'pre_post_nota_ind' : a boolean to indicate whether election held in pre or post nota period
#see pre_processing_data_meta.txt for more details on the rationale of this code
gen_state_combine['pre_post_nota_ind'] = 100000000000
gen_state_combine.loc[(gen_state_combine['Year'] <= 2012), 'pre_post_nota_ind'] = 0
gen_state_combine.loc[(gen_state_combine['Year'] >= 2014), 'pre_post_nota_ind'] = 1

#Excluding Nagaland(13), Tripura(16), Meghalaya(17) and Karnataka(29)
gen_state_combine.loc[((gen_state_combine['Year'] == 2013) & (gen_state_combine['state_code_new'].isin([13,16,17,29])) & (gen_state_combine['gen/state'] == 1)), 'pre_post_nota_ind'] = 0
#Including Delhi(7), Rajasthan(8), Mizoram(15), Chhatisgarh(22), Madhya Pradesh(23)
gen_state_combine.loc[(gen_state_combine['Year'] == 2013) & (gen_state_combine['state_code_new'].isin([7,8,15,22,23])) & (gen_state_combine['gen/state'] == 1), 'pre_post_nota_ind'] = 1
```
